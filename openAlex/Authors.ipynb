```python
import requests
import pandas as pd
import numpy as np

response_API = requests.get('https://api.openalex.org')
response_API.status_code # 200 means success
```

```python
url_with_cursor = 'https://api.openalex.org/authors?filter=last_known_institution.id:https://openalex.org/I201448701&cursor={}'
cursor = '*'
count=1
row = []
# loop through pages
while cursor:
    url = url_with_cursor.format(cursor)
    print("\n"+ str(count) + ": " + url)
    page_with_results = requests.get(url).json()
    count+=1
    for a in page_with_results['results']: 
        if 'mag' in a['ids']:
            magid = a['ids']['mag']
        else:
            magid = np.nan

        row.append([magid, np.nan, a['display_name'], a['display_name'].lower(), 
                    a['last_known_institution']['id'], 
                    a['works_count'], a['cited_by_count'], a['orcid']])
    cursor = page_with_results['meta']['next_cursor']
Author_df = pd.DataFrame(row, columns=["Author_ID", "Rank", "Display_name", 
                                       "Normalized_name", "Last_known_affiliation_ID", 
                                       "Paper_count", "Citation_count", "Orcid"])

Author_df
```
