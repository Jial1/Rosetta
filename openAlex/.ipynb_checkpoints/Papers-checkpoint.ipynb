```python
UW_works_url_with_cursor = 'https://api.openalex.org/works?filter=institutions.id:I201448701&cursor={}'

cursor = '*'

count = 1

rows = []


# loop through pages
while cursor:
    
    UW_works_url = UW_works_url_with_cursor.format(cursor)
    print("\n" + str(count) + ": " + UW_works_url)

    UW_works = requests.get(UW_works_url).json()
        
    
    for r in UW_works['results']:
        
        if 'mag' in r['ids']:
            magid = r['ids']['mag']
        else:
            magid = np.nan
        if 'primary_location' not in r or r['primary_location'] is None or 'source' not in r['primary_location'] or r['primary_location']['source'] is None:
            display_name = np.nan
        else:
            display_name = r['primary_location']['source']['display_name']
        
        rows.append([magid, r['title'], r['publication_year'], 
                     r['publication_date'], r['doi'], display_name, np.nan, np.nan])

    # update cursor to meta.next_cursor
    cursor = UW_works['meta']['next_cursor']
    count += 1


UW_works_df = pd.DataFrame(rows, columns=["Paper_id", "title", "publication_year", 
                                          "publication_date", "DOI", "Journal_name",
                                          "Conference_series_ID", "language"])
UW_works_df
```
